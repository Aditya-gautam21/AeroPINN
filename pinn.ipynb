```python
import pyvista as pv
import numpy as np
from pathlib import Path

class DataLoader:
    def load_simulations(folder):
        internal_file = list(Path(folder).glob("*internal.vtu"))[0]

        mesh = pv.read(internal_file)

        points = mesh.points

        x = points[:,0]
        y = points[:,1]

        velocity = mesh["U"]
        pressure = mesh["p"]

        u = velocity[:,0]
        v = velocity[:,1]

        X = np.stack([x,y],axis=1)
        Y = np.stack([u,v,pressure],axis=1)

        return X,Y
```

```python
loader = DataLoader()
loader.load_simulations("Data")
```
